### Q1. Write a python program to display all the header tags from wikipedia.org

```python
page=requests.get("https://en.wikipedia.org/wiki/Main_Page")
soup=BeautifulSoup(page.content)
soup.title.text
```

### Q3. Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) and make data frame.

```python
url = "https://www.imdb.com/list/ls091520106/"
page2 = requests.get(url)

soup = BeautifulSoup(page2.content)
soup.title.text
```

```python
name = soup.find_all("h3", class_="lister-item-header")

movies_name = []
for i in name:
    for j in i.find_all("a"):
        movies_name.append(j.text)


year = soup.find_all("span",class_="lister-item-year text-muted unbold")
year_of_release = []
for k in year:
    a=k.text.replace('(','')
    year_of_release.append(a.replace(')','')) 


rating = soup.find_all("div",class_="ipl-rating-star small")


IMDB_rating = [] 
for i in rating:
      IMDB_rating.append(float(i.text))
        
# Make data frame of top 100 movies on IMDB
IMDB_top_100=pd.DataFrame({})
IMDB_top_100['movies_name']=movies_name
IMDB_top_100['year_of_release']=year_of_release
IMDB_top_100['IMDB_rating']=IMDB_rating  
IMDB_top_100
```

### Q5. Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape:
a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.


```python
url = "https://www.icc-cricket.com/rankings/mens/team-rankings/odi"
page5 = requests.get(url)

soup5 = BeautifulSoup(page5.content)
soup5.title.text
```

```python
team = soup5.find_all("span",class_='u-hide-phablet')
team_name = []
for i in team:
    team_name.append(i.text)
matches = []
points = []
ratings = []
new_list = []

for i in soup5.find_all("td",class_='rankings-block__banner--matches'):
    matches.append(i.text)
for i in soup5.find_all("td",class_='rankings-block__banner--points'):
    points.append(i.text)
for i in soup5.find_all("td",class_='rankings-block__banner--rating u-text-right'):
    ratings.append(i.text.replace("\n",""))
for i in soup5.find_all("td",class_='table-body__cell u-center-text'):
    new_list.append(i.text)
for i in range(0,len(new_list)-1,2):
    matches.append(new_list[i])
    points.append(new_list[i+1])
for i in soup5.find_all("td",class_='table-body__cell u-text-right rating'):
    ratings.append(i.text)
    
# Make data frame of top 10 ICC teams
icc=pd.DataFrame({})
icc['Team_name']=team_name[:10]
icc['Matches']=matches[:10]
icc['Points']=points[:10]
icc['Ratings']=ratings[:10]
icc    
```

### Q9. Write a python program to scrape mentioned details from dineout.co.in :
i) Restaurant name

ii) Cuisine

iii) Location

iv) Ratings

v) Image URL

```python
page = requests.get("https://www.dineout.co.in/delhi-restaurants/buffet-special")

soup13=BeautifulSoup(page.content)
soup13.title.text
```

```python
name=[]
for i in soup13.find_all("div", class_="restnt-info cursor"):
    name.append(i.text)
    
location=[]
for i in soup13.find_all("div", class_="restnt-loc ellipsis"):
    location.append(i.text)

cuisine = []
for i in soup13.find_all("span", class_="double-line-ellipsis"):
    cuisine.append(i.text.split('|')[1])

rating=[]
for i in soup13.find_all("div", class_="restnt-rating rating-3"):
    rating.append(i.text)
    
for i in soup13.find_all("div", class_="restnt-rating rating-4"):
    rating.append(i.text)

images = []
for i in soup13.find_all("img", class_="no-img"):
    images.append(i['data-src'])
```

```python
# Make data frame
DineOut=pd.DataFrame({})
DineOut['Restaurant Name']=name
DineOut['Location']=location
DineOut['Cuisine']=cuisine  
DineOut['Rating']=rating  
DineOut['IMAGES']=images
DineOut
```

### Q10. Write a python program to scrape the details of top publications from Google Scholar from https://scholar.google.com/citations?view_op=top_venues&hl=en
i) Rank

ii) Publication

iii) h5-index

iv) h5-median

```python
url = "https://scholar.google.com/citations?view_op=top_venues&hl=en"
page15 = requests.get(url)

soup15 = BeautifulSoup(page15.content)
soup15.title.text
```

```python
rank = []
for i in soup15.find_all('td', class_="gsc_mvt_p"):
    rank.append(i.text)
    
publication = []
for i in soup15.find_all('td',class_="gsc_mvt_t"):
    publication.append(i.text)
    
h5_index = []
for i in soup15.find_all('a',class_="gs_ibl gsc_mp_anchor"):
    h5_index.append(i.text)
    
h5_median = []
for i in soup15.find_all('span', class_="gs_ibl gsc_mp_anchor"):
    h5_median.append(i.text)
    
    
#Create Dataframe
df = pd.DataFrame({'Rank':rank, 'Publication':publication, 'H5-index': h5_index, 'H5-median': h5_median})
df
```
